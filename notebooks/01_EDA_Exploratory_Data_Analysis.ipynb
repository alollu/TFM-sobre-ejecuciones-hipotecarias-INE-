```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# --- 1. CARGA DE DATOS ---
# Ajustamos la ruta para buscar en la carpeta 'data' que está un nivel arriba
df = pd.read_csv('../data/datos_historicos.csv')

# Configuración de estilo
sns.set_theme(style="whitegrid")
plt.figure(figsize=(20, 15))

# --- 2. GRÁFICO TEMPORAL ---
plt.subplot(2, 2, 1)
sns.lineplot(data=df, x='periodo', y='total', hue='titular', marker='o')
plt.title('Evolución de Ejecuciones Hipotecarias')

# --- 3. GRÁFICO POR CCAA ---
plt.subplot(2, 2, 2)
top_ccaa = df.groupby('com_nom')['total'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_ccaa.values, y=top_ccaa.index, palette='viridis')
plt.title('Top 10 Comunidades Autónomas')

# --- 4. DETECCIÓN DE OUTLIERS ---
plt.subplot(2, 2, 3)
sns.boxplot(x='total', y='titular', data=df)
plt.title('Distribución y Valores Atípicos')

# --- 5. TENDENCIA GENERAL ---
plt.subplot(2, 2, 4)
df_total = df.groupby('periodo')['total'].sum().reset_index()
sns.barplot(data=df_total, x='periodo', y='total', color='skyblue')
plt.title('Total Ejecuciones por Año')

plt.tight_layout()
plt.show()
print("✅ EDA Completado.")
```
